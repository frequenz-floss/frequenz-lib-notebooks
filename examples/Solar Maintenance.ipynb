```

from dateutil.parser import parse as _deepnote_parse
start_timestamp = _deepnote_parse('2025-01-01T00:00:00.000Z').date()
```

```
language = 'English'
```

```
resample_period = '3600'
```

```
rolling_view_duration = 30
```

```
baseline_models = ['7-day MA']
```

```
# parse input block values
start_timestamp = datetime.datetime.combine(
    start_timestamp, datetime.datetime.min.time()
).astimezone(datetime.timezone.utc)
end_timestamp = datetime.datetime.now().astimezone(datetime.timezone.utc)
rolling_view_duration = min(
    rolling_view_duration, (end_timestamp - start_timestamp).days
)
language = {"English": "en", "Deutsch": "de"}.get(language, "en")

client_site_info = {
    "latitude": getattr(configs[microgrid_id].meta, "latitude", 0.0),
    "longitude": getattr(configs[microgrid_id].meta, "longitude", 0.0),
    "altitude": getattr(configs[microgrid_id].meta, "altitude", 0.0),
    "peak_power_watts": sum(
        pv.peak_power
        for pv in getattr(configs[microgrid_id], "pv", {}).values()
    ),
    "rated_power_watts": sum(
        pv.rated_power
        for pv in getattr(configs[microgrid_id], "pv", {}).values()
    ),
    "efficiency": 0.85,
}

USER_REQUEST = {
    "time_zone": time_zone,
    "start_timestamp": start_timestamp,
    "end_timestamp": end_timestamp,
    "rolling_view_duration": rolling_view_duration,
    "language": language,
    "weather_service_address": os.environ["WEATHER_API_URL"],
    "reporting_service_address": os.environ["REPORTING_API_URL"],
    "microgrid_ids": [int(microgrid_id)],
    "component_ids": [
        configs[microgrid_id].component_type_ids(
            component_type="pv", component_category=component_category
        )
    ],
    "client_site_info": [client_site_info],
    "baseline_models": baseline_models,
    "large_resample_period_seconds": int(resample_period),
}
```

```
# --- for visualisation in deepnote --- #
mid = int(microgrid_id)
production_table_view = plot_data.production_table_view
rolling_view_real_time = plot_data.real_time_view[mid]
rolling_view_short_term = plot_data.rolling_view_short_term[mid]
# --- hack to make the deepnote plot work --- #
import numpy as np

if "predictions_7-day sampled MA" not in rolling_view_short_term.columns:
    rolling_view_short_term["predictions_7-day sampled MA"] = np.nan
if "predictions_weather-based-forecast" not in rolling_view_short_term.columns:
    rolling_view_short_term["predictions_weather-based-forecast"] = np.nan
if "predictions_simulation" not in rolling_view_short_term.columns:
    rolling_view_short_term["predictions_simulation"] = np.nan
# ---------- #
rolling_view_long_term = plot_data.rolling_view_long_term[mid]
# --- hack to make the deepnote plot work --- #
if "predictions_7-day MA" not in rolling_view_long_term.columns:
    rolling_view_long_term["predictions_7-day MA"] = np.nan
if "predictions_weather-based-forecast" not in rolling_view_long_term.columns:
    rolling_view_long_term["predictions_weather-based-forecast"] = np.nan
if "predictions_simulation" not in rolling_view_long_term.columns:
    rolling_view_long_term["predictions_simulation"] = np.nan
# ---------- #
rolling_view_average = plot_data.rolling_view_average[mid]
df_daily_production = plot_data.daily_production[mid]
statistical_view_dict = plot_data.statistical_profiles[mid]
if "grouped" in statistical_view_dict:
    statistical_view_grouped = statistical_view_dict["grouped"]
if "continuous" in statistical_view_dict:
    statistical_view_continuous = statistical_view_dict["continuous"]
if "24h_continuous" in statistical_view_dict:
    statistical_view_24h_continuous = statistical_view_dict["24h_continuous"]

# automate the real-time view plot on deepnote
import pandas as pd

_microgrid_info_cids = [
    str(x)
    for x in configs[microgrid_id].component_type_ids(
        component_type="pv", component_category=component_category
    )
]
components_list = [
    k for k in rolling_view_real_time.keys() if k in _microgrid_info_cids
]
if not rolling_view_real_time.empty:
    hack_rolling_view_real_time = pd.melt(
        rolling_view_real_time,
        id_vars=["x-axis"],
        value_vars=components_list,
        var_name="Component",
        value_name="Power (kW)",
    )
else:
    hack_rolling_view_real_time = pd.DataFrame(
        columns=["x-axis", "Component", "Power (kW)"]
    )
# ---------------------------------- #
```
