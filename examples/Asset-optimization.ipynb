```python
import pandas as pd    
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import os
from dotenv import load_dotenv


import frequenz.lib.notebooks.reporting.asset_optimization as ao 

import sys
from frequenz.data.microgrid import MicrogridData 

plt.rcParams["figure.figsize"] = (30,6.66)
plt.rcParams["axes.grid"] = True
```

```python
SERVICE_ADDRESS = os.environ["REPORTING_API_URL"]
API_KEY = os.environ["REPORTING_API_KEY"]

mdata = MicrogridData(
    server_url=SERVICE_ADDRESS,
    key=API_KEY, 
    microgrid_config_path="./microgrids.toml",
)

mids = [i.replace("iot", "") for i in mdata.microgrid_ids] 
```

# Power flow

```python
microgrid_id = '13'
```

```python
mid = int(microgrid_id)
ctypes = mdata.microgrid_configs[str(mid)].component_types()
```

```python
component_types = ['grid','pv','battery','consumption','chp']
```

```python

from dateutil.parser import parse as _deepnote_parse
end_date = _deepnote_parse('2025-05-28T00:00:00.000Z').date()
```

```python
days_back = '7'
```

```python
resolution = '900'
```

```python
end_dt = datetime(end_date.year, end_date.month, end_date.day)
start_dt = end_dt - timedelta(days=int(days_back))
```

```python
data = await ao.fetch_data(
    mdata=mdata,
    component_types=component_types,
    mid=mid,
    start_time=start_dt,
    end_time=end_dt,
    resampling_period=timedelta(seconds=int(resolution)), 
    fetch_soc=True,
)

df = data
```

```python
ao.plot_power_flow(df)
```

```python
ao.plot_battery_power(df)
```

```python
ao.plot_power_flow_trade(df)
```

```python
df[["grid", "pv", "chp", "battery"]].plot(color=["grey", "gold", "cornflowerblue", "red"]);
```
