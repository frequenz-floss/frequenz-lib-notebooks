# Imports

```
import logging
import os
from datetime import UTC, datetime, timedelta

import pandas as pd
from frequenz.client.common.metric import Metric
from frequenz.client.common.microgrid.components import ComponentStateCode
from frequenz.client.reporting import ReportingApiClient

from frequenz.gridpool import MicrogridConfig
from frequenz.lib.notebooks.alerts.alert_email import (
    AlertEmailConfig,
    ExportOptions,
    compute_time_since,
    generate_alert_email,
    generate_no_alerts_email,
    plot_alerts,
)
from frequenz.lib.notebooks.notification_service import EmailConfig, EmailNotification
from frequenz.lib.notebooks.notification_utils import (
    format_email_preview,
    send_test_email,
    validate_email_config,
)
from frequenz.lib.notebooks.reporting.state_analysis import (
    fetch_and_extract_state_durations,
)

logging.basicConfig(
    level=logging.WARNING,
    format="%(asctime)s %(levelname)s %(name)s: %(message)s",
    datefmt="%H:%M:%S",
)
logging.getLogger("frequenz.lib.notebooks").setLevel(logging.WARNING)
```

```
configs: dict[str, "MicrogridConfig"] = MicrogridConfig.load_configs(
    microgrid_config_dir="/work"
)
available_microgrids = sorted(list(int(x) for x in configs.keys()))

assets = {"battery", "pv", "ev"}
available_component_types = sorted(
    {t for config in configs.values() for t in config.component_types() if t in assets}
)

available_component_states = [state.name for state in ComponentStateCode]
```

```
filter_no_alerts = 'True'
```

```
default_email_config = {
    "smtp_server": smtp_server,  # replace with your SMTP server
    "smtp_port": smtp_port,  # replace with your SMTP port
    "smtp_user": smtp_user,  # replace with your SMTP user
    "smtp_password": smtp_password,  # replace with your SMTP password
    "from_email": sender,  # sender email
    "recipients": recipients.replace(" ", "").split(","),  # accepts multiple recipients
    "subject": email_subject if email_subject else "Component Error",  # email subject
    "message": "Empty message. Should be replaced automatically.",  # email body
}

# Bypasses __post_init__ for validation-only use
config = EmailConfig.__new__(EmailConfig)

# Set attributes manually
for k, v in default_email_config.items():
    setattr(config, k, v)

errors = validate_email_config(config, check_connectivity=True)
if errors:
    try:
        from IPython.display import display

        display(errors)
    except ImportError:
        print("Validation errors:", *errors, sep="\n")
else:
    config.subject = "Test Email - Alert Notebook Notification Service Configuration"
    success = send_test_email(config)
```

```
# parse user input and set up some defaults and other parameters
microgrid_components_request = []
metrics_to_request = [Metric.AC_ACTIVE_POWER]
for microgrid_id, microgrid_config in configs.items():
    if microgrid_id not in microgrid_ids:
        continue
    for ctype in component_types:
        if ctype not in microgrid_config.component_types():
            continue
        inverters = microgrid_config.component_type_ids(
            component_type=ctype, component_category="inverter"
        )
        components = microgrid_config.component_type_ids(
            component_type=ctype, component_category="component"
        )
        if ctype == "battery":
            metrics_to_request.append(Metric.DC_POWER)
        logging.info(
            "Microgrid %s: Found %s inverters (%s) and %s components (%s) of type `%s`",
            microgrid_id,
            len(inverters) if inverters else 0,
            inverters,
            len(components) if components else 0,
            components,
            ctype,
        )
        cid_request = []
        cid_request = cid_request + inverters if inverters else cid_request
        cid_request = cid_request + components if components else cid_request
        if cid_request:
            microgrid_components_request += [(int(microgrid_id), cid_request)]
            logging.info("Microgrid request: %s", microgrid_components_request)
        else:
            logging.warning("No components found for microgrid %s", microgrid_id)

# define the start time, end time, and, resampling period for the data request
end_time = datetime.now().astimezone(UTC)
start_time = end_time - timedelta(minutes=float(lookback_duration_minutes))
resampling_period_seconds = timedelta(seconds=1)

# convert to boolean
sort_by_severity = sort_by_severity == "True" if sort_by_severity else False
group_by_component = group_by_component == "True" if sort_by_severity else False
filter_no_alerts = filter_no_alerts == "True" if sort_by_severity else True

# whether to include warnings (errors are included by default)
include_warnings = True

# set up alert conditions (to be) checked
conditions_checked = ["All error codes"]
conditions_checked += ["All warning codes"] if include_warnings else []
conditions_checked += [f"ComponentState_{state}" for state in component_states_to_alert]
```

```
# set up the Reporting API client
client = ReportingApiClient(
    server_url=os.environ["REPORTING_API_URL"],
    auth_key=os.environ["API_KEY"],
    sign_secret=os.environ["API_SECRET"],
)

# fetch data and extract state durations and alert records.
all_states, alert_records = await fetch_and_extract_state_durations(
    client=client,
    microgrid_components=microgrid_components_request,
    metrics=metrics_to_request,
    start_time=start_time,
    end_time=end_time,
    resampling_period=resampling_period_seconds,
    alert_states=[ComponentStateCode[name] for name in component_states_to_alert],
    include_warnings=include_warnings,
)
```

```
# create dataframes
all_states_df = pd.DataFrame(all_states)
alert_records_df = pd.DataFrame(alert_records)

# add more stats (time since start, time since end)
if not all_states_df.empty:
    all_states_df["time_since_start"] = all_states_df.apply(
        lambda row: compute_time_since(row, "start_time"), axis=1
    )
    all_states_df["time_since_end"] = all_states_df.apply(
        lambda row: compute_time_since(row, "end_time"), axis=1
    )

if not alert_records_df.empty:
    alert_records_df["time_since_start"] = alert_records_df.apply(
        lambda row: compute_time_since(row, "start_time"), axis=1
    )
    alert_records_df["time_since_end"] = alert_records_df.apply(
        lambda row: compute_time_since(row, "end_time"), axis=1
    )
```

```
# variables are for display purposes
subject = "Preview"
alert_file_name = ""

# set up the config for generating alert emails
alert_email_config = AlertEmailConfig(
    notebook_url=notebook_url,
    displayed_rows=int(rows_to_display) if rows_to_display else 20,
    sort_by_severity=sort_by_severity,
    group_by_component=group_by_component,
    filter_no_alerts=filter_no_alerts,
)

email_config = EmailConfig.from_dict(default_email_config)

if not alert_records_df.empty:  # send an alert email
    email_config.subject += f" - {len(alert_records_df)} alerts detected"
    subject = email_config.subject

    # generate HTML email body with alert details
    html_body = generate_alert_email(
        alert_records=alert_records_df,
        config=alert_email_config,
        checks=conditions_checked,
    )
    email_config.message = html_body

    # optionally add attachments (a list of files)
    email_config.attachments = None
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    alert_file_name = f"alert_details_{timestamp}.csv"
    alert_records_df.to_csv(alert_file_name, index=False)
    email_config.attachments = [alert_file_name]

    # create a visualisation of the alert records
    img_path = plot_alerts(
        records=alert_records_df,
        plot_type="all",
        export_options=ExportOptions(
            # format="png",
            show=True,
        ),
    )
    email_config.attachments += img_path if img_path else []

    # send email
    email_notification = EmailNotification(config=email_config)
    email_notification.send()
else:  # send a no alerts email
    email_config.subject = "No Alerts Found"
    email_config.message = generate_no_alerts_email(
        checks=conditions_checked, notebook_url=notebook_url
    )
    email_notification = EmailNotification(config=email_config)
    email_notification.send()
```
